**#Task 1: Calculate Factorial Using a Function**

```python
def factorial(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

# Sample call
num = 5
print(f"The factorial of {num} is: {factorial(num)}")
```

**Task 2: Math Module for Calculations**


```python

import math

def calculate_math_functions():
  """
  Asks user for a number, calculates and displays its square root, natural log, and sine.
  """
  try:
    num_str = input("Enter a number: ")
    num = float(num_str)

    # Calculate and display results
    sqrt_num = math.sqrt(num)
    print(f"Square root: {sqrt_num}")

    if num > 0:
      log_num = math.log(num)
      print(f"Natural logarithm (base e): {log_num}")
    else:
      print("Natural logarithm is undefined for non-positive numbers.")

    sin_num = math.sin(num)
    print(f"Sine (in radians): {sin_num}")

  except ValueError:
    print("Invalid input. Please enter a valid number.")

# Call the function
calculate_math_functions()
```
