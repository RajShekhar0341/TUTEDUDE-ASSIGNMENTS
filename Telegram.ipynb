```python
from telethon.sync import TelegramClient
from telethon.errors.rpcerrorlist import PeerFloodError, UserPrivacyRestrictedError, UserBotError
from telethon.tl.functions.channels import InviteToChannelRequest
import pandas as pd
import traceback
import time
import random

api_id = 1219108
api_hash = '6a62fccfd576a92a04e6f021bd5c8bf2'
phone = '+917382140605'
session_name = 'add1'
channel_username = '@xooxbettingline'  # Use only the username, not the full URL

client = TelegramClient(session_name, api_id, api_hash)

# Connect to Telegram and sign in
client.connect()
if not client.is_user_authorized():
    client.send_code_request(phone)
    client.sign_in(phone, input('Enter the code: '))

# Load user data from CSV using pandas
file_path = '/content/members.csv'
df = pd.read_csv(file_path)

# Get user input for the range
startfrom = int(input("Start From = "))
endto = int(input("End To = "))

n = 0

for index, row in df.iterrows():
    # Convert user_id and access_hash to integers if they are not already
    try:
        user_id = int(row['user id'])
        access_hash = int(row['access hash'])
    except ValueError:
        print(f"Skipping row {index} due to invalid user id or access hash.")
        continue

    if startfrom <= index <= endto:
        n += 1
        if n % 50 == 0:
            print("Sleeping for 15 minutes...")
            time.sleep(900)

        try:
            if pd.notna(row['username']):
                print(f"Adding {row['username']} (ID: {user_id})")
                client(InviteToChannelRequest(
                    channel_username,
                    [row['username']]
                ))
                print("Waiting for 60-180 seconds...")
                time.sleep(random.randint(60, 130))
            else:
                print(f"No username provided for ID {user_id}, moving to the next user.")
        except PeerFloodError:
            print("Getting Flood Error from Telegram. Script is stopping now. Please try again after some time.")
            break
        except UserPrivacyRestrictedError:
            print("The user's privacy settings do not allow you to do this. Skipping.")
        except UserBotError:
            print("Bot error: Ensure the bot is an admin in the channel.")
            break
        except Exception:
            traceback.print_exc()
            print("Unexpected Error")
    elif index > endto:
        print("Members added successfully")
        break
```
